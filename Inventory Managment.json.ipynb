```python
import json
import time

fd = open('Inventory.json','r')
js = fd.read()
fd.close()

record = json.loads(js)
         
print("----------------------Menu--------------------------")

for key in record.keys() :
    print(key ,":",record[key]["Name"],"\t|",record[key]["Price"] ,"\t|", record[key]["Qn"])
print("-----------------------------------------------------")
ur_name = input("Enter your Name : ")
ur_mail = input("Enter the mail Id : ")
ur_phone = input("Enter the phone number : ")
ur_id = str(input("Enter the Id of the product : "))
ur_qn = int(input("Enter the qunatity of the product : "))

if(int(record[ur_id]["Qn"]) >= ur_qn):
    print("-------------------------------------------")
    print("Item Name               : ",record[ur_id]["Name"])
    print("Item Price              : ",record[ur_id]["Price"],'Rs')
    print("Quantity                : ",ur_qn)
    print("Gst                     : ",'13%')
    print("-------------------------------------------")
    print("Price                   : ",int(record[ur_id]["Price"]) * ur_qn,'Rs')
    price = int(record[ur_id]["Price"]) * ur_qn
    gst = (int(record[ur_id]["Price"]) * ur_qn) * 13/100
    if (price >= 5000):
        print("Discount                 : ",(int(record[ur_id]["Price"]) * ur_qn)*10/100)
        Discount = (int(record[ur_id]["Price"]) * ur_qn)*10/100
        Total_Amount = price-Discount+gst
        print("Total Amount             : ",price-Discount+gst,'Rs')
        sale = ur_name +", "+ur_phone+", "+ur_mail+", "+record[ur_id]["Name"]+", "+str(ur_qn)+", "+str(Total_Amount)+time.ctime()
    else:
            Total_Amount=price+gst
            print("Total Amount            : ",price+gst,'Rs')
            sale = ur_name +", "+ur_phone+", "+ur_mail+", "+record[ur_id]["Name"]+", "+str(ur_qn)+", "+str(Total_Amount)+time.ctime()
    
    record[ur_id]["Qn"]=record[ur_id]["Qn"]-ur_qn
else:
    print("We don't have the specific amount : ")
    print("We can provide ",record[ur_id]["Qn"],"if you want to purchase then press Y or N ")
    print("Press Y/N ")
    ch = input("Enter Y or N")
    if (ch == 'Y' or ch =='y'):
        print("-------------------------------------------")
        print("Item Name               : ",record[ur_id]["Name"])
        print("Item Price              : ",record[ur_id]["Price"],'Rs')
        print("Quantity                : ",ur_qn)
        print("-------------------------------------------")
        print("Price                   : ",int(record[ur_id]["Price"]) * ur_qn,'Rs')
        price = int(record[ur_id]["Price"]) * ur_qn
        if (price >= 5000):
            print("Discount             : ",(int(record[ur_id]["Price"]) * ur_qn)*10/100)
            Discount = (int(record[ur_id]["Price"]) * ur_qn)*10/100
            gst = (int(record[ur_id]["Price"]) * ur_qn) * 12/100
            Total_Amount = price-Discount+gst
            print("Total Amount         : ",price-Discount+gst,'Rs')
            sale = ur_name +", "+ur_phone+", "+ur_mail+", "+record[ur_id]["Name"]+", "+str(ur_qn)+", "+str(Total_Amount)+time.ctime()
        record[ur_id]["Qn"]='0'
    
    else:
        print("Thank You :)")
    

js = json.dumps(record)
fd = open("Inventory.json",'w')
fd.write(js)
fd.close()

fd = open('Sales.json','w')
fd.write(sale)
fd.close()
```
